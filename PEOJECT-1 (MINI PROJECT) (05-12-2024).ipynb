```python
#TASK 02 - CHECKING MISSING VALUES & IF THERE ARE SOME DUPLICATE ROWS OR NOT
import pandas as pd

df = pd.read_csv("/content/sample_data/dataset.csv") #Read the Dataset file
duplicates = df[df.duplicated()] #Find duplicate rows
missing = df.isnull().sum() #Find rows with missing values

# Print or save the results
print("Duplicate Rows:")
print(duplicates)

print("\nRows with Missing Values:")
print(missing)
```

```python
#TASK 03 - PLOTTING THE DISTRIBUTION OF VARIOUS DISCRETE FEATURES ON SEASON/HOLIDAY/WEEKEND/WEATHERCODE
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/sample_data/dataset.csv") #Read the Dataset file
df['timestamp'] = pd.to_datetime(df['timestamp']) #Convert timestamp to datetime objects
features = ['season', 'is_holiday', 'is_weekend', 'weather_code'] #Create discrete features for plotting

plt.figure(figsize=(12, 10)) #Setting up the plot size
#Loop through each feature and plot the distribution
for idx, feature in enumerate(features, 1):
    plt.subplot(2, 2, idx)  #2x2 grid of subplots
    sns.countplot(x=feature, data=df)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.yticks(rotation=45)

plt.tight_layout() #Adjust layout to prevent overlap
plt.show()
```

```python
#TASK 04 - LOOKING AT THE DATA TYPES OF EACH VARIABLE, TRANSFORM TIMESTAMP IN TYPE & SET IT AS INDEX
import pandas as pd

df = pd.read_csv("/content/sample_data/dataset.csv") #Read the Dataset file

print(df.dtypes) #Printing Datatypes
print("-------------------------------------------")
df['timestamp'] = pd.to_datetime(df['timestamp']) #Convert 'timestamp' column to datetime
if df.index.name != 'timestamp': #Check if 'timestamp' is already the index
      df.set_index('timestamp', inplace=True)
      print("Column 'timestamp' is now an index.")
else:
    print("Column 'timestamp' is already an index.")
print("-------------------------------------------")
df
```

```python
#TASK 05 - MAKING FEATURE ENGINEERING. EXTRACTING NEW COLUMNS (DAY OF WEEK, DAY OF MONTH, HOUR, MONTH, SEASON, YEAR ETC.)
import pandas as pd

df = pd.read_csv("/content/sample_data/dataset.csv") #Read the Dataset file
df['timestamp'] = pd.to_datetime(df['timestamp']) #Convert 'timestamp' column to datetime
df['year'] = df['timestamp'].dt.year #Extracting 'year' column
df['month'] = df['timestamp'].dt.month #Extracting 'month' column
df['hour'] = df['timestamp'].dt.hour #Extracting 'hour' column
df['day_of_week'] = df['timestamp'].dt.day_of_week #Extracting 'day_of_week' column
df['day_of_month'] = df['timestamp'].dt.days_in_month #Extracting 'day_of_month' column

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

df['season_name'] = df['month'].apply(get_season) #Extracting 'season_name' column
df
```

```python
#TASK 06 - VISUALIZING THE CORRELATION WITH A HEATMAP
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("/content/sample_data/dataset.csv") #Read the Dataset file
corr_columns = df.select_dtypes(include=[np.number]).columns #Exclude non-numeric columns
matrixcorr = df[corr_columns].corr() #Calculate the correlation matrix

#Correlation Map
plt.figure(figsize=(12,10))
sns.heatmap(matrixcorr,annot=True,cmap='Blues',fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
#TASK 07 - VISUALIZING THE CORRELATION OF THE TARGET VARIABLE & THE OTHER FEATURES WITH BARPLOT
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("/content/sample_data/dataset.csv") #Read the Dataset file
targetvariable = 'hum' #Target Variable
corr_columns = df.select_dtypes(include=[np.number]).columns #Exclude non-numeric columns
matrixcorr = df[corr_columns].corr()[targetvariable].drop(targetvariable) #Calculate the correlation matrix

#Bar Plot Graph
plt.figure(figsize=(12,6))
matrixcorr.plot(kind = 'bar')
plt.title('Correlation with Target Variable')
plt.xlabel('Features')
plt.ylabel('Correlation with Humidity')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
#TASK 08 - PLOTTING BIKE SHARES OVER TIME BY USING LINEPLOT
#Considering Count column values as Bike shares
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/sample_data/dataset.csv") #Read the Dataset file

#Line Plot Graph
plt.figure(figsize=(16,8))
plt.plot(df.index, df['cnt'])
plt.title('Bike Shares Over Time')
plt.xlabel('Time')
plt.ylabel('Bike Shares')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
#TASK 09 - PLOTTING BIKE SHARES BY MONTHS & YEAR_OF_MONTH (USE LINEPLOT, POINTPOT, BARPLOT)
#Considering Count column values as Bike shares
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/sample_data/dataset.csv") #Read the Dataset file
df['year_month'] = pd.to_datetime(df['timestamp']).dt.to_period('M') #Extracting 'year_month' column
bikeshares = df.groupby('year_month')['cnt'].sum() #Grouping the bike shares values

#Line Plot Graph
plt.figure(figsize=(14,6))
plt.title('Bike Shares by Month and Year-Month')
sns.lineplot(x=bikeshares.index.astype('string'), y=bikeshares)
plt.xlabel('Month')
plt.ylabel('Bike Shares')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()
plt.show()

#Point Plot Graph
plt.figure(figsize=(14,6))
plt.title('Bike Shares by Month and Year-Month')
sns.pointplot(x=bikeshares.index.astype('string'), y=bikeshares)
plt.xlabel('Month')
plt.ylabel('Bike Shares')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()
plt.show()

#Bar Plot Graph
plt.figure(figsize=(14,6))
plt.title('Bike Shares by Month and Year-Month')
sns.barplot(x=bikeshares.index.astype('string'), y=bikeshares)
plt.xlabel('Month')
plt.ylabel('Bike Shares')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#TASK 10 - PLOTTING BIKE SHARES BY HOURS ON (HOLIDAYS, WEEKEND, SEASON)
#Considering Count column values as Bike shares
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/sample_data/dataset.csv") #Read the Dataset file
df['hour'] = pd.to_datetime(df['timestamp']).dt.hour #Extracting 'hour' column

#Line Plot Graph for Bike Shares on Holidays
plt.figure(figsize=(12,6))
plt.title('Bike Shares on Holidays')
sns.lineplot(x=df['hour'], y=df['cnt'], data=df[df['is_holiday'] == 1])
plt.xlabel('Hour')
plt.ylabel('Bike Shares')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()

#Line Plot Graph for Bike Shares on Weekends
plt.figure(figsize=(12,6))
plt.title('Bike Shares on Weekends')
sns.lineplot(x=df['hour'], y=df['cnt'], data=df[df['is_weekend'] == 1])
plt.xlabel('Hour')
plt.ylabel('Bike Shares on Weekends')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()

#Line Plot Graph for Bike Shares by Season
plt.figure(figsize=(12, 6))
plt.title('Bike Shares by Season')
sns.lineplot(x=df['hour'], y=df['cnt'],  hue=df['season'], data=df)
plt.xlabel('Hour')
plt.ylabel('Bike Shares by Season')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
#TASK 12 - PLOTTING BIKE SHARES BY DAY OF MONTH
#Considering Count column values as Bike shares
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/sample_data/dataset.csv") #Read the Dataset file
df['day_of_month'] = pd.to_datetime(df['timestamp']).dt.day #Extracting 'day_of_month' column
bikeshares = df.groupby('day_of_month')['cnt'].mean()

#Bar Plot Graph for Bike Shares by Day of Month
plt.figure(figsize=(12, 6))
plt.title('Bike Shares by Day of Month')
bikeshares.plot(kind='bar', color="skyblue")
plt.xlabel('Hour')
plt.ylabel('Bike Shares')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
#TASK 13 - PLOTTING BIKE SHARES BY YEAR & PLOTTING BIKE SHARES ON HOLIDAYS BY SEASON
#Considering Count column values as Bike shares
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/sample_data/dataset.csv") #Read the Dataset file
df['year'] = pd.to_datetime(df['timestamp']).dt.year #Extracting 'year' column
holidaydata = df[df['is_holiday'] == 1]

#Bar Plot Graph for Bike Shares by year
plt.figure(figsize=(12, 6))
plt.title('Bike Shares by Year')
sns.barplot(x=df['year'], y=df['cnt'], color='yellow' )
plt.xlabel('Hour')
plt.ylabel('Bike Shares')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()

#Line Plot Graph for Bike Shares on holidays by season
plt.figure(figsize=(12, 6))
plt.title('Bike Shares by Season')
sns.lineplot(x=holidaydata['season'], y=holidaydata['cnt'], color='skyblue')
plt.xlabel('Hour')
plt.ylabel('Bike Shares')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
#TASK 14 - VISUALIZING THE DISTRIBUTION OF BIKE SHARES BY WEEKDAY-WEEKEND WITH USING PIECHART & BARPLOT
#Considering Count column values as Bike shares
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/sample_data/dataset.csv") #Read the Dataset file
weekenddata = df[df['is_weekend']==1]
weekdaydata = df[df['is_weekend']==0]
bikeshares = [weekenddata['cnt'].sum(), weekdaydata['cnt'].sum()]
colors=['yellow', 'skyblue']
label = ['Weekend', 'Weekday']

#Pie Chart
plt.figure(figsize=(6, 6))
plt.title('Bike Shares Weekend vs Weekdays')
plt.pie(bikeshares, labels=label, startangle=90, colors=colors, autopct='%1.1f%%')
plt.tight_layout()
plt.show()

#Bar Plot Graph
plt.figure(figsize=(6, 6))
plt.title('Bike Shares Weekend vs Weekdays')
plt.bar(label, bikeshares, color=colors)
plt.tight_layout()
plt.show()
```
